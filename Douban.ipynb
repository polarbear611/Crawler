```python
import requests, re
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
url = 'https://movie.douban.com/top250'
headers={
        'User-Agent':'Mozilla/5.0 (Windows NT 6.3; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.94 Safari/537.36',
        }
ranking = 0
for start in range(0, 250, 25):
    params['start'] = start
    print("Getting movie informations of ranking {} to {} ...".format(start + 1, start + 25))
    response = requests.get(url, params=params)
    print("Getting movie informations of ranking {} to {} succeed.".format(start + 1, start + 25))
    text = response.text
    soup = BeautifulSoup(text, "lxml")
    movie_infos = soup.find_all('li')
    grid_view = soup.find('ol', class_='grid_view')
    movie_items = grid_view.find_all('li')
    print("Parsing movie informations ...")
    for movie in movie_items:
        ranking += 1
        print("Ranking {}：".format(ranking))
        try:
            titles = movie.find_all('span', class_='title')
            title = titles[0].text
            try:
                orginal_title = titles[1].text.split('/')[1].strip()
            except IndexError:
                orginal_title = ''
            other_titles = movie.find('span', class_='other').text.split('/')[1].strip()
            #导演、主演、年代、国家、分类
            descriptions = movie.find('p', class_='').text
            director, stars = descriptions.split('\n')[1].strip().split('\xa0\xa0\xa0')
            director = director[3:].strip()
            stars = stars[3:].strip().replace(',', '/')
            #print (director, stars)
            year, country, genres = descriptions.split('\n')[2].strip().split('/')
            #print(year.strip(), country.strip(), genres.strip())
            rating = movie.find('span', class_='rating_num').text
            rating_times = movie.find_all('span')[-2].text[:-3]
            inq = movie.find('span', class_='inq').text

            #print(title, orginal_title, other_titles, director, stars, year.strip(), country.strip(), genres.strip(), rating, rating_times, inq)
            print("title: {}\n orginal_title:{}\n other_titles:{}\n director:{}\n stars:{}\n year:{}\n country:{}\n genres:{}\n rating:{}\n rating_times:{}\n inq".format(title, orginal_title, other_titles, director, stars, year.strip(), country.strip(), genres.strip(), rating, rating_times, inq))
            with open("../Data-Analysis/WebXML/movie_info_top250.csv", 'a') as f:
                f.write(','.join([str(ranking), title, orginal_title, other_titles, director, stars, year.strip(), country.strip(), genres.strip(), rating, rating_times, inq]) + '\n')
        except Exception as e:
            print(e)
        
    seconds = randint(1, 3)
    print("Sleep {} seconds for a pause...".format(seconds))
    sleep(seconds)
```
